## How are property transaction values distributed?

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
df = pd.read_csv(r"E:\Project\Japan_Housing_Prices\data\cleaned\cleaned.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
plt.figure(figsize=(10,6))
y = df["TotalTransactionValue"]
plt.title("Value of Properties")
plt.xlabel("No of Properties")
plt.ylabel("Value(in yen)")
plt.hist(y, bins=20)
plt.show()
```

## Distribution of Property Transaction Values

### Observation

- Most property transaction values are concentrated in the lower price ranges.
- The distribution is highly right-skewed.
- A small number of extremely expensive properties create a long right tail.
- These extreme values stretch the x-axis, causing the majority of transactions to appear compressed in the lower price range.

# Does Residental Properties have an effect by area?

```python
plt.figure(figsize=(10,6))
plt.title("Properties by Area")
plt.xlabel("No of Properties")
plt.ylabel("Area")
x = df["Area"]
plt.hist(x)
plt.show()
```

### Observation

- Most residential properties have an area between approximately 50 m² and 250 m².
- The distribution is positively skewed, indicating that smaller and medium-sized properties are much more common than larger ones.
- The number of properties decreases as area increases.
- A small number of exceptionally large properties create a long right tail, suggesting the presence of outliers.

# Do larger properties generally sell for higher prices?

```python
plt.figure(figsize=(10,6))
x = df["Area"]
y = df["TotalTransactionValue"]
plt.title("Properties by their Prices")
plt.xlabel("Area( in square meters)")
plt.ylabel("Value ( in YEN )")
plt.scatter(x,y,
            color="red",
            alpha=0.5)
plt.show()
```

### Observation

- Most properties are concentrated in the lower ranges of both area and transaction value.
- Larger properties generally tend to have higher transaction values, suggesting a positive relationship between the two variables.
- The spread of transaction values increases as property area increases, indicating greater price variability among larger properties.
- A small number of extremely high-value transactions appear as outliers.

 # Does being farther from the nearest station affect the property's transaction value?

```python
plt.figure(figsize=(10,6))
plt.title("Value for Train Proximity")
plt.xlabel("Time( in minutes )")
plt.ylabel("Value")
x = df["AverageTimeToStation"]
y = df["TotalTransactionValue"]
plt.scatter(x,y,
            color="red",
            alpha=0.3)
plt.show()
```

### Observation

- Most properties are located within approximately 30 minutes of the nearest train station.
- Properties closer to stations exhibit a wide range of transaction values, including many of the highest-priced transactions.
- As the distance from the station increases, the number of observations decreases considerably.
- A weak negative relationship appears to exist between station distance and transaction value, although other factors clearly influence property prices.
